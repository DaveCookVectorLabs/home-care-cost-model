# Home Care Cost Model — Analysis

Companion notebook to the working paper. Reproduces the three headline findings from 5,000 synthetic scenarios and the cross-province subsidy gap table.

All data is deterministic and under CC BY 4.0. Code is MIT-licensed. This is a reference framework, not clinical or financial advice.

```python
import pandas as pd
import sys
from pathlib import Path

datasets = Path('../datasets')
scenarios = pd.read_csv(datasets / 'home_care_scenarios.csv')
archetypes = pd.read_csv(datasets / 'home_care_cost_model_archetypes.csv')
subsidy_gap = pd.read_csv(datasets / 'home_care_subsidy_gap.csv')
tax_sensitivity = pd.read_csv(datasets / 'home_care_tax_relief_sensitivity.csv')

print(f'Scenarios: {len(scenarios):,}')
print(f'Archetypes: {len(archetypes):,}')
print(f'Subsidy gap: {len(subsidy_gap):,}')
print(f'Tax sensitivity: {len(tax_sensitivity):,}')
```

## 1. Hybrid savings vs all-PSW counterfactual

```python
hybrid = scenarios[scenarios['recommended_housekeeping_hours_per_week'] > 0].copy()
hybrid['savings_pct'] = hybrid['hybrid_savings_vs_all_psw_monthly_cad'] / hybrid['all_psw_cost_comparison_monthly_cad'] * 100
print(hybrid['savings_pct'].describe().round(2))
```

## 2. Tax relief effectiveness by income band

```python
print(tax_sensitivity.groupby('net_family_income_cad')[['total_credits_value_cad','effective_credit_fraction_of_oop']].mean().round(3))
```

## 3. Cross-province subsidy gap

```python
print(subsidy_gap.groupby('jurisdiction_code')[['hours_gap_per_week','dollar_gap_monthly_cad']].mean().round(2).sort_values('dollar_gap_monthly_cad', ascending=False))
```
